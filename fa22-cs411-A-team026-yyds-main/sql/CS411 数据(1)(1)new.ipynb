```python
def team(team_language,team_num,team_member,team_founder,team_foundtime,team_lowrank,team_highrank):
    n=70
    for i in range(n):
        foundtime = '"' + team_foundtime[i] + '"'
        print('(',i,',',
              foundtime,',',
              team_founder[i],',',
              team_num[i],',',
              team_language[i],',',
              team_lowrank[i],',',
              team_highrank[i],
              ')',',',
             )
        
        print("\n")
```

```python
def skin(skin_WeaponType,skin_MarketPrice,skin_Credit_Suggest,skin_SkinName):
    n=1024
    for i in range(n):
        WeaponType = '"' + skin_WeaponType[i] + '"'
        SkinName = '"' + skin_SkinName[i] + '"'
        print('(',i,',',
              WeaponType,',',
              skin_MarketPrice[i],',',
              int(skin_Credit_Suggest[i]+0.5),',',
              SkinName,')',',',
             )
        
```
